# Example 1: Before and After Message

```python
def my_decorator(func):

    def wrapper():

        print("Before Function")

        func()

        print("After Function")

    return wrapper
```

```python
@my_decorator
def greet():

    print("Hello")
```

```python
greet()
```

# Example 2: Login Check

```python
def login_required(func):

    def wrapper():

        logged_in = True

        if logged_in:
            func()
        else:
            print("Please Login")

    return wrapper
```

```python
@login_required
def dashboard():

    print("Welcome to Dashboard")
```

```python
dashboard()
```

# Example 3: Function Call Counter

```python
def counter(func):

    def wrapper():

        print("Function Called")

        func()

    return wrapper
```

```python
@counter
def display():

    print("Displaying Data")
```

```python
display()
```

# Example 4: Execution Timer

```python
import time

def timer(func):

    def wrapper():

        start = time.time()

        func()

        end = time.time()

        print("Time:", end - start)

    return wrapper
```

```python
@timer
def process():

    time.sleep(2)

    print("Processing...")
```

```python
process()
```

# Example 5: Repeat Function Multiple Times

```python
def repeat(func):

    def wrapper():

        for i in range(3):
            func()

    return wrapper
```

```python
@repeat
def hello():

    print("Hello")
```

```python
hello()
```
