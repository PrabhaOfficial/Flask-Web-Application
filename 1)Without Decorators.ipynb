# Without Decorators

```python
def add():

    print("Function Started")

    print("Adding Numbers")

    print("Function Ended")


def subtract():

    print("Function Started")

    print("Subtracting Numbers")

    print("Function Ended")


def multiply():

    print("Function Started")

    print("Multiplying Numbers")

    print("Function Ended")


add()
subtract()
multiply()
```

# With Decorators

```python
def logger(func):

    def wrapper():

        print("Function Started")

        func()

        print("Function Ended")

    return wrapper
```

```python
@logger
def add():

    print("Adding Numbers")


@logger
def subtract():

    print("Subtracting Numbers")


@logger
def multiply():

    print("Multiplying Numbers")
```

```python
add()
subtract()
multiply()
```
